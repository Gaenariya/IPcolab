```python
img = cv2.imread('/content/gdrive/My Drive/Images/girl.jpg')
print(img.shape)
gImg1 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gImg1, cmap='gray')
```

```python
gMul = np.zeros((gImg1.shape[0], gImg1.shape[1]), dtype=np.uint8)
```

```python
def saturation(val):
  if val > 255:
    return 255
  elif val < 0:
    return 0
  else:
    return val
```

```python
val = 2

for h in range(gImg1.shape[0]):
  for w in range(gImg1.shape[1]):
    mulVal = np.int32(gImg1[h, w]) + val
    gMul[h, w] = saturation(gImg1[h, w] * val)
```

```python
val = 0.2

for h in range(gImg1.shape[0]):
  for w in range(gImg1.shape[1]):
    mulVal = np.int32(gImg1[h, w]) * val
    gMul[h, w] = saturation(mulVal)

plt.imshow(gMul, cmap='gray', vmin = 0, vmax = 255)
```
